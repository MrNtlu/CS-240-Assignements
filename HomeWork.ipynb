```python
from __future__ import print_function, division
import thinkstats2
import numpy as np
import pandas as pd
import nsfg
import thinkplot
```

```python
resp = nsfg.ReadFemResp()
```

```python
resp.columns
```

```python
f1=resp.head(20)
f2=resp.tail(30)
frames=[f1,f2]
result=pd.concat(frames)
result
```

```python
columnLength=len(result.columns)
rowLength=result.shape[0]
```

```python
print ("There are ", columnLength," columns and ", rowLength, "rows.")
```

```python
print ("Oldest:",result["agescrn"].max(), " Youngest:", result["agescrn"].min())
```

```python
under25=result[result["agescrn"]<=25]
under25
```

```python
avarageArray=under25["numpregs"].values
total=0
for i in range(len(avarageArray)):
    total+=avarageArray[i]
print (total/len(avarageArray), "is the avarage of pregnancies in lifetime")
```

```python
histTable=result["fmarno"]
hist=thinkstats2.Hist(histTable, label="Marriage Frequency")
thinkplot.Hist(hist)
thinkplot.Config(xlabel="Marriage(times)",ylabel="Age")
```

```python
neverMarried=result[result["fmarno"]<1]
married=result[result["fmarno"]>=1]
nMhist=thinkstats2.Hist(neverMarried["totincr"], label="NonMarried")
mhist=thinkstats2.Hist(married["totincr"], label="Married")
```

```python
width = 0.4
thinkplot.PrePlot(2)
thinkplot.Hist(nMhist, align='right', width=width)
thinkplot.Hist(mhist, align='left', width=width)
thinkplot.Config(xlabel='weeks', ylabel='Count')
```

```python
minMar=married["totincr"].min()
maxMar=married["totincr"].max()

minNever=neverMarried["totincr"].min()
maxNever=neverMarried["totincr"].max()

meanNever=married["totincr"].mean()
meanMarried=neverMarried["totincr"].mean()

stdNever=married["totincr"].std()
stdMarried=neverMarried["totincr"].std()

varNever=married["totincr"].var()
varMarried=neverMarried["totincr"].var()

print ("Never Married Min:", minNever, "Max:", maxNever, "Mean:", meanNever, "Standar Deviation:", stdNever, "Variance:",varNever,)
```

```python
print ("Married Min:", minMar, "Max:", maxMar, "Mean:", meanMarried, "Standar Deviation:", stdMarried, "Variance:",varMarried,)
```
